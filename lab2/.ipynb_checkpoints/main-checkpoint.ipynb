```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

## Завантажимо картинки

```python
books = cv2.imread('./books.jpeg', 0)

book = cv2.imread('./book.jpeg', 0)
```

```python
def show_img(img):
  converted = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
  plt.imshow(converted)
```

```python
show_img(books)
```

```python
show_img(book)
```

```python
def daisy_kp(img):
    detector = cv2.FastFeatureDetector_create()
    daisy = cv2.xfeatures2d.DAISY_create()

    kps = detector.detect(img)

    return daisy.compute(img, kps)
```

```python
def cv2_BFM(kp1, des1, kp2, des2, img1, img2):
    bf = cv2.BFMatcher(cv2.NORM_L1)
    matches = bf.match(des1,des2)

    best_matches = sorted(matches, key=lambda x:x.distance)

    output_image = cv2.drawMatches(img1, kp1, img2, kp2, best_matches[:20], None, flags=2)

    return output_image
```

```python
def custom_BFM(kp1, des1, kp2, des2, img1, img2):
    matches = []

    for i, k1 in enumerate(des1):
        min_distance = 1000
        min_index = 0

        for j, k2 in enumerate(des2):
            distance = np.sum(np.abs(k1 - k2))

            if(distance < min_distance):
              min_distance = distance
              min_index = j

        matches.append(cv2.DMatch(_distance=min_distance, _imgIdx=0, _queryIdx=i, _trainIdx=min_index))

    best_matches = sorted(matches, key=lambda x:x.distance)

    output_image = cv2.drawMatches(img1, kp1, img2, kp2, best_matches[:20], None, flags=2)

    return output_image
```

```python
kp1, desc1 = daisy_kp(book)
kp2, desc2 = daisy_kp(books)
```

```python
res1 = cv2_BFM(kp1, desc1, kp2, desc2, book, books)
show_img(res1)
```

```python
res2 = custom_BFM(kp1, desc1, kp2, desc2, book, books)
show_img(res2)
```
